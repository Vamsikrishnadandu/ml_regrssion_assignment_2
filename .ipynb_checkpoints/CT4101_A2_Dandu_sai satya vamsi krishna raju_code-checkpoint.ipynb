```python
import pandas as pd
```

```python
data = pd.read_csv("steel.csv")
data
```

```python
data.describe()
```

```python
data.info()   ##we can see that there are no null values across all the features and target varible
```

```python
Y = data['tensile_strength']
Y
```

```python
#dividing the into X and Y, to split the data for training and testing the model accordingly.

X = data.drop("tensile_strength", axis = 1)
X   ## Predictors, we have to train the model.
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

Decision Tree Implementaion, by cross validation:

For cross validation, the error metrics i selected are as follows:
Mean sqaured error
Mean absolute errr

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_validate

dt_default = DecisionTreeRegressor()

##corss validation: 

cv_results_dt_default = cross_validate(dt_default, X, Y, cv = 10, scoring = {'mse': 'neg_mean_squared_error',
        'mae': 'neg_mean_absolute_error'}, return_train_score = True, return_estimator = True)

cv_results_dt_default

##The domain independent error score i selected is mean squared error:
##average training MSE for 10 folds, from the cross_validation we did above:
print(f"Training MSE: {-cv_results_dt_default['train_mse'].mean():.4f}")

#average testing MSE for 10 folds:
print(f"Tetsing MSE: {-cv_results_dt_default['test_mse'].mean():.4f}")

##Domain specific error measure i choose is mean absolute error:
##Avg training MAE:
print(f"Training MAE: {-cv_results_dt_default['train_mse'].mean():.4f}")

##Average testing MAE:
print(f"Testing MAE: {-cv_results_dt_default['test_mae'].mean():.4f}")
```

```python
dt_default
```

```python
cv_results_dt_default
```

```python
##hyperparamter tuning for the above default decision tree model we created:
##I am using GridSearchCV to find the best parameters, for my model:

from sklearn.model_selection import GridSearchCV

param_grid_dt = {"max_depth": [10,15,20,25,30], "min_samples_split": [2,5,10,20]}

grid_search_dt = GridSearchCV(estimator = DecisionTreeRegressor(), param_grid = param_grid_dt, cv = 10,
            scoring = "neg_mean_squared_error", n_jobs = -1, return_train_score = True )

grid_search_dt.fit(X,Y)
```

```python
best_dt = grid_search_dt.best_estimator_

##now doing the cross validation to the new new estimator, we found after hyperparamter tuning:

cv_dt_tuned = cross_validate(
    best_dt, X, Y, cv=10,
    scoring={
        'mse': 'neg_mean_squared_error',
        'mae': 'neg_mean_absolute_error'
    },
    return_train_score=True
)
cv_dt_tuned
```

```python
##Results of the cross_validation, after tuning the hyperparamters of decision tree model. As we can see thet best parameters above which we got based on grid_search

##average training MSE after tuning:
print(f"Training MSE: {-cv_dt_tuned['train_mse'].mean():.4f}")

#average testing MSE for 10 folds:
print(f"Tetsing MSE: {-cv_dt_tuned['test_mse'].mean():.4f}")

##Domain specific error measure i choose is mean absolute error:

##Avg training MAE:
print(f"Training MAE: {-cv_dt_tuned['train_mae'].mean():.4f}")

##Average testing MAE:
print(f"Testing MAE: {-cv_dt_tuned['test_mae'].mean():.4f}")
```

```python
##Feature sclaing using Standardsclaer:
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled   ##values 
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn_default = KNeighborsRegressor()
knn_default

cv_knn_default = cross_validate(knn_default,X_scaled, Y, scoring = {'mse' : "neg_mean_squared_error",
                                                           'mae':"neg_mean_absolute_error"}, cv = 10, n_jobs = -1, return_train_score = True, return_estimator = True)
cv_knn_default

##showing the results of cross_validation, while using knn default model.

print("Below are the reults of cross_validation for the default KNN model:")

print(f"Training MSE: {-cv_knn_default['train_mse'].mean():.4f}")

#average testing MSE for 10 folds:
print(f"Tetsing MSE: {-cv_knn_default['test_mse'].mean():.4f}")

##Domain specific error measure i choose is mean absolute error:

##Avg training MAE:
print(f"Training MAE: {-cv_knn_default['train_mae'].mean():.4f}")

##Average testing MAE:
print(f"Testing MAE: {-cv_knn_default['test_mae'].mean():.4f}")
```

```python
cv_knn_default
```

```python
knn_default  ##this is the default model we have used above, just wanted see the exact default parameter values used..
```

```python
##Using grid search to find the best params for the knn model as part of the hyperparameter tuning:
param_grid_knn = {"n_neighbors" : [3,5,7,9,11,15,21], "weights" : ["uniform", "distance"], "metric" : ["euclidean", "manhattan"]}
grid_search_knn = GridSearchCV(estimator = KNeighborsRegressor(), param_grid = param_grid_knn,cv = 10,scoring = "neg_mean_squared_error",n_jobs = -1, return_train_score = True)

grid_search_knn.fit(X_scaled,Y)
```

```python
grid_search_knn.best_params_
```

```python
best_knn = grid_search_knn.best_estimator_

cv_knn_tuned = cross_validate(best_knn, X, Y,cv = 10, scoring = {"mse":"neg_mean_squared_error", "mae":"neg_mean_absolute_error"},
                             return_train_score = True)
cv_knn_tuned
```

```python

print("Below are the reults of cross_validation after tuning the KNN model:")
##after tuning KNN model:

print(f"Training MSE: {cv_knn_tuned['train_mse'].mean():.4f}")

#average testing MSE for 10 folds:
print(f"Tetsing MSE: {-cv_knn_tuned['test_mse'].mean():.4f}")

##Domain specific error measure i choose is mean absolute error:

##Avg training MAE:
print(f"Training MAE: {cv_knn_tuned['train_mae'].mean():.4f}")

##Average testing MAE:
print(f"Testing MAE: {-cv_knn_tuned['test_mae'].mean():.4f}")
```
